```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set()
```

```python
def intensity(x):
    return x
```

```python
h = 0.01
M = 100

x=[1]
y = []

for m in range(1, 5*M):
    y_new = np.random.choice([0,1], p=[1 - h, h])
    y.append(y_new)

Y = np.cumsum(y)
Y
```

```python
for m in range(1, M):
    argum = 0
    for n in range(m):
        argum += intensity(x[n])*h
    print(argum)
    x_new = x[0] + Y[int(argum/h)]
    x.append(x_new)

x
```

```python
x = np.vstack(x)
x = x[:, 0]
```
